```python
import numpy as np
import matplotlib.pyplot as plt

h = [3, 12, 5, 18, 45]
b = ('A', 'B', 'C', 'D', 'E')
x = np.arange(len(b))
plt.bar(x, h)
plt.xticks(x, b)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
plt.bar(
    ['a', 'b', 'ab', 'o'],
    [4, 4.5 ,3 ,2],
    label='Smoking Rate',
    align='center'
)

plt.title('Smoking Rate with Blood Group')
plt.xlabel('Blood Group')
plt.ylabel('Smoking Rate')
plt.legend()
```

```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
plt.bar(
    ['a', 'b', 'ab', 'o'],
    [5, 4.5 ,3 ,4],
    label='Smoking Rate in 2019',
    color='g',
    align='center',
    alpha=0.5
)

plt.bar(
    ['a', 'b', 'ab', 'o'],
    [6, 5.5 ,4 ,3],
    label='Smoking Rate in 2020',
    color='m',
    align='center',
    alpha=0.5
)
plt.title('Smoking Rate with Blood Group')
plt.xlabel('Blood Group')
plt.ylabel('Smoking Rate')
plt.legend()
```

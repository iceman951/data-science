```python
import matplotlib.pyplot as plt
labels = ['The cats', 'Les Miserables', 'Phantom of the Opera', 'Evita', 'West Side Story', 'Others']

Rate = [80, 70, 60, 50, 60, 20]

plt.pie(Rate, labels=labels, autopct='%1.1f%%')
plt.title('Popular Musical in Thailand - 2024')
plt.show()
```

```python
import matplotlib.pyplot as plt
labels = ['The cats', 'Les Miserables', 'Phantom of the Opera', 'Evita', 'West Side Story', 'Others']

Rate = [80, 70, 60, 50, 60, 20]
ee = (0.0, 0.3, 0, 0, 0, 0.1)

plt.pie(Rate, 
        labels=labels,
        explode=ee,
        autopct='%1.1f%%'
        )
plt.title('Popular Musical in Thailand - 2024')
plt.legend(labels)
plt.show()
```

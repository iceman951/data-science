```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

h = [[160], [165], [170], [173], [180]]
w = [[60], [65], [72], [75], [80]]

plt.scatter(h, w)
plt.title('Relationshion of high and weight of 5 students')
plt.xlabel('Student high')
plt.ylabel('Student weight')

model = LinearRegression()
model.fit(h, w)

w = model.predict([[175]])
print(w)

plt.plot(h, model.predict(h))
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

h = [[1.58], [1.62], [1.69], [1.76], [1.82]]
w = [[58], [63], [72], [73], [85]]

model = LinearRegression()
model.fit(h, w)

plt.scatter(h, w)
plt.title('Relationshion of high and weight of 5 students')
plt.xlabel('Student high')
plt.ylabel('Student weight')

plt.plot(h, model.predict(h))

RSS = np.sum((np.ravel(w) - np.ravel(model.predict(h)))**2)
print("RSS: %.2f"% RSS)
print("R-squared: %.2f"% model.score(h, w))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures

degree = 2
Polynomial_features = PolynomialFeatures(degree=degree)

h = [[1.5],[2],[3],[4],[5],[6]]
w = [[1.5],[2.5],[4],[4],[4.5],[5]]

h_poly = Polynomial_features.fit_transform(h)
print(h_poly)

model = LinearRegression()

model.fit(h_poly, w)
w_poly_pred = model.predict(h_poly)

plt.scatter(h, w, s=10)

plt.plot(h, w_poly_pred, color='m')
plt.show()

print('R-Squared for training set: %.4f %', model.score(h_poly,w))
```
